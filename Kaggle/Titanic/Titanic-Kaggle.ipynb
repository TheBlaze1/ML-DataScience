```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.dropna(inplace=True) # droping rows with missing data
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
df.drop("Name", axis =1, inplace = True)
df.drop("PassengerId", axis = 1, inplace = True)

X = df.drop("Survived", axis = 1)
y = df["Survived"]
categorical_features = ["Sex","Cabin","Ticket","Embarked"]
one_hot= OneHotEncoder()
transformer = ColumnTransformer([('one_hot',one_hot,categorical_features)], remainder = "passthrough")
transformed_X = transformer.fit_transform(X) 
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(transformed_X,y, test_size=0.2)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test) 
```

```python
import numpy as np
# Improve the loaded and retrained model to match the previous dataset
# Trying different ammounts of estimators
np.random.seed(40)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)
    print(f"Model accuracy on the test set: {clf.score(X_test,y_test) * 100:.2f}")
    print("")
```

#### The best score after evalution is with 90 estimators and is matching the current model score - that is 75.68
